```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lung_ct = nib.load("/content/drive/MyDrive/Medical_Imaging/03-Data-Formats/03-Preprocessing/lung_043.nii.gz")
lung_ct_data = lung_ct.get_fdata()
```

```python
lung_ct_data_standart = lung_ct_data/3071
```

```python
plt.figure()
plt.imshow(np.rot90(lung_ct_data_standart[:,:,50]),cmap='bone')
```

```python
lung_ct_lung_window = np.clip(lung_ct_data,-1000,-500)
plt.figure()
plt.imshow(np.rot90(lung_ct_lung_window[:,:,50]),cmap='bone')
```

```python
lung_ct_soft_tissue_window = np.clip(lung_ct_data,-250,250)
plt.figure()
plt.imshow(np.rot90(lung_ct_soft_tissue_window[:,:,50]),cmap='bone')
```
